```python
from unsloth import FastLanguageModel

max_seq_length = 2048

dtype = None

load_in_4bit = True

# ëª¨ë¸ ì„¤ì •
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="./llama3-inst",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
    cache_dir='/data'
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    use_gradient_checkpointing = True,
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

```python
import pandas as pd
from datasets import Dataset

df = pd.read_csv('./mbti_datasets_1.csv')

dataset = Dataset.from_pandas(df)

dataset
```

```python
def formatting_mbti_prompts_func(examples):
    output_texts = []
    for i in range(len(examples['text'])):
        messages = [
            {"role": "system", "content": "ì •í™•í•œ ì±—ë´‡ìœ¼ë¡œì„œ ìƒëŒ€ë°©ì˜ ì…ë ¥ì— ëŒ€í•´ MBTIë¥¼ ë§ì¶”ì. ëª¨ë“  ëŒ€ë‹µì€ MBTI 16ê°œì¤‘ í•˜ë‚˜ë¡œ ëŒ€ë‹µí•´ì¤˜."},
            {"role": "user", "content": "{}".format(examples['text'][i])},
            {"role": "assistant", "content": "{}".format(examples['mbti'][i])}
        ]
        # LLAMAì˜ ëª¨ë¸ í´ë˜ìŠ¤ì— ë§ëŠ” ê¸°ë³¸ ì±„íŒ… í…œí”Œë¦¿ ì ìš©
        chat_message = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=False)
        output_texts.append(chat_message)

    return {"text": output_texts}
```

```python
formatted_dataset = dataset.map(formatting_mbti_prompts_func, batched=True)

split_dataset = formatted_dataset.train_test_split(test_size=0.2, seed=42)

# ë¶„í• ëœ ë°ì´í„°ì…‹ì—ì„œ í›ˆë ¨ê³¼ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ ì¶”ì¶œ
train_dataset = split_dataset['train']
test_dataset = split_dataset['test']
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
import torch

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    dataset_num_proc=10,  # ë©€í‹°í”„ë¡œì„¸ì‹±ì„ ìœ„í•´ ë” ë§ì€ í”„ë¡œì„¸ìŠ¤ë¥¼ ì‚¬ìš© (ì½”ë©ì—ì„œ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ë¦¬ì†ŒìŠ¤ì— ë”°ë¼ ì¡°ì •)
    packing=False,  # ì§§ì€ ì‹œí€€ìŠ¤ì˜ ê²½ìš° Falseê°€ ë” ë‚˜ì„ ìˆ˜ ìˆìŒ
    args=TrainingArguments(
        per_device_train_batch_size=2,  # GPU ë©”ëª¨ë¦¬ì— ë§ì¶° ë°°ì¹˜ í¬ê¸° ì¡°ì •
        gradient_accumulation_steps=4,  # ë°°ì¹˜ í¬ê¸°ë¥¼ íš¨ê³¼ì ìœ¼ë¡œ ëŠ˜ë¦¬ê¸° ìœ„í•´ ê·¸ë˜ë””ì–¸íŠ¸ ëˆ„ì  ì‚¬ìš©
        warmup_steps=100,  # ì›Œë°ì—… ìŠ¤í…ì„ ëŠ˜ë ¤ ëª¨ë¸ì´ ë” ì•ˆì •ì ìœ¼ë¡œ í•™ìŠµí•˜ë„ë¡ ì„¤ì •
        max_steps=500,  # í•™ìŠµ ìŠ¤í… ìˆ˜ë¥¼ ëŠ˜ë ¤ ì¶©ë¶„í•œ í•™ìŠµ ì‹œê°„ í™•ë³´
        learning_rate=5e-5,  # ë³´í¸ì ìœ¼ë¡œ ì‚¬ìš©ë˜ëŠ” í•™ìŠµë¥ 
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps=50,  # ë¡œê·¸ë¥¼ ëœ ë¹ˆë²ˆí•˜ê²Œ ì¶œë ¥í•˜ì—¬ í•™ìŠµ ì†ë„ì— ì˜í–¥ì„ ì£¼ì§€ ì•Šë„ë¡ í•¨
        optim="adamw_torch",  # ìµœì‹  ì˜µí‹°ë§ˆì´ì € ì‚¬ìš© (PyTorchì˜ ê¸°ë³¸ AdamW êµ¬í˜„)
        weight_decay=0.01,  # ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©ë˜ëŠ” weight decay
        lr_scheduler_type="cosine",  # í•™ìŠµë¥ ì´ ë¶€ë“œëŸ½ê²Œ ê°ì†Œí•˜ë„ë¡ cosine ìŠ¤ì¼€ì¤„ ì‚¬ìš©
        seed=3407,
        output_dir="outputs",
        save_steps=100,  # ì²´í¬í¬ì¸íŠ¸ë¥¼ ìì£¼ ì €ì¥í•˜ì—¬ ì¤‘ê°„ì— í•™ìŠµì„ ì¤‘ë‹¨í•˜ê³ ë„ ì¬ê°œ ê°€ëŠ¥í•˜ë„ë¡ ì„¤ì •
    ),
)

trainer_stats = trainer.train()
```
