```python
import pandas as pd
import pickle
```

```python
with open("data1.pkl", 'rb') as file:
    data = pickle.load(file)

data
```

```python
df = pd.DataFrame(data, columns=['text', 'senti'])
df
```

```python
mood_mapping = {
    '0': "공포",
    '1': "놀람",
    '2': "분노",
    '3': "슬픔",
    '4': "중립",
    '5': "행복",
    '6': "혐오"
}
```

```python
df['senti'] = df['senti'].map(mood_mapping)
```

```python
df
```

```python
mood_percentages = df['senti'].value_counts(normalize=True) * 100
mood_percentages = mood_percentages.round(2)

mood_percentages
```

```python
df.to_csv('senti.csv', index=False)
```
